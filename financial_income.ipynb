
# Income&Expense_py

## Import pandas

```python
import pandas as pd
import numpy as np
```

## Archives Getting

archive .XLSX

the month must be exactly the sheet name you want into analysis 

```python
month = input('What month: ')
```

```python
df_items = pd.read_excel('items_de_receitas&despesas.xlsx', sheet_name = month, skiprows = 0)
```

## Data Treatment

### Renaming column names 

```python
df_items.head()
```

```python
df_items = df_items.rename(columns = {'Descrição': 'descricao','Valor':'valor', 'Data':'data', 'Tipo':'tipo', 'Classificação':'classif'})
```

```python
df_items.head()
```

### Reordering columns

```python
 df_items = df_items[['descricao','data','tipo','classif', 'valor']]
```

### Verifying column type

```python
df_items.dtypes
```

### Deleting column

```python
df_items = df_items.drop(['data'], axis = 'columns')
```

```python
df_items.head()
```

### Grouping data

Grouping by 'descricao' and suming their respective values

```python
df_items = df_items.groupby(['descricao', 'tipo','classif']).agg({'valor':'sum'}).reset_index(drop = False)
```

```python
df_items
```

### Sorting Data

This is the sorting order I like to sort my items. You are able to modify it by changing the numbers in the dict beneath ;)

```python
order_sort = {'Despesa':1, 'Receita':0, 'Fixa':2, 'Variável':3, 'Extra':4, 'Eventual Esperada':5, 'Emergẽncia':6, 'Dívida':7} 
```

```python
df_items = df_items.sort_values(['valor'], ascending = True)
```

```python
df_items = df_items.sort_values(by=['tipo', 'classif'], key=lambda x: x.map(order_sort)).reset_index(drop = True)
```

```python
df_items
```

## Business Rule

let's check your income and expense of the month :)

### Grouping by 'type' and by 'classif' and cumulating the sum

```python
df_items['valor_acum'] = df_items.groupby(['tipo', 'classif']).cumsum()
```

```python
df_items
```

### Calculating Final Balance

```python
df_final = df_items.groupby(['tipo']).agg({'valor':'sum'}).reset_index(drop = False)
```

```python
df_final
```

```python
df_final = df_final.sort_values(by=['tipo'], key=lambda x: x.map(order_sort)).reset_index(drop = True)
```

```python
df_final
```

```python
balance = round(df_final['valor'][0] - df_final['valor'][1], 2)
```

```python
balance
```

```python
df_balance = pd.DataFrame({'tipo':['Saldo'], 'valor':balance})
```

```python
df_balance
```

### Concating dataframes

```python
df_final = pd.concat([df_final, df_balance]).reset_index(drop = True)
```

```python
df_final
```

## Exporting data

```python
 title = 'receitas&despesas_' + month + '.xlsx'
```

```python
writer = pd.ExcelWriter(title)
```

### 'df_items' to .XLSX

```python

df_items.to_excel(writer, sheet_name = 'items', index = False)
```

### 'df_final' to .XLSX

```python
df_final.to_excel(writer, sheet_name = 'Saldo_final', index = False)
```

```python
writer.save()
```
